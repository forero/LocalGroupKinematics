```python
pec_vel_boxes = {}
box_id = 0
delta_box = 200.0
min_box = -1000.0
for i in range(10):
    for j in range(10):
        for k in range(10):
            min_x = min_box + i*delta_box
            min_y = min_box + j*delta_box
            min_z = min_box + k*delta_box
            kk = (x>min_x) & (x<(min_x+delta_box))
            #print(min_x, (min_x+delta_box))
            kk &= (y>min_y) & (y<(min_y+delta_box))            
            kk &= (z>min_z) & (z<(min_z+delta_box))
            #print(np.count_nonzero(kk))
            pec_vel_boxes[box_id] = vel_data_0[kk,0]
            #print(len(pec_vel_boxes[box_id]))
            box_id +=1
```

```python
for i in range(1000):
    pec_vel = pec_vel_boxes[i]
    n_v = len(pec_vel)
    plt.plot(np.sort(pec_vel), np.linspace(0., 1.0, n_v), color='black', alpha=0.01)

n_t = len(vel_data_300)
plt.plot(np.sort(vel_data_300[:,0]), np.linspace(0,1,n_t), color='red')
plt.xlim(0,1000)
plt.ylim(0,1)
```

```python
tan_vel_boxes = {}
box_id = 0
delta_box = 200.0
min_box = -1000.0
for i in range(10):
    for j in range(10):
        for k in range(10):
            min_x = min_box + i*delta_box
            min_y = min_box + j*delta_box
            min_z = min_box + k*delta_box
            kk = (x>min_x) & (x<(min_x+delta_box))
            #print(min_x, (min_x+delta_box))
            kk &= (y>min_y) & (y<(min_y+delta_box))            
            kk &= (z>min_z) & (z<(min_z+delta_box))
            #print(np.count_nonzero(kk))
            tan_vel_boxes[box_id] = vel_data_0[kk,1]
            #print(len(pec_vel_boxes[box_id]))
            box_id +=1
```
