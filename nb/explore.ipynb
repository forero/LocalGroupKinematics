```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = '../data/pairs_TNG300-1.hdf5'
data = {}
f = h5py.File(filename, 'r')
for k in f.keys():
    data[k] = f[k][...]
f.close()
print('finished reading {}'.format(filename))
print(len(data['N_A']))
```

```python
plt.scatter(data['vmax_A'], np.log10(data['N_A']), alpha=1.0, s=1.0)
plt.scatter(data['vmax_B'], np.log10(data['N_B']), alpha=1.0, s=1.0)
```

```python
plt.scatter(data['vmax_A'], data['vmax_B'], alpha=1.0, s=1.0)
```
