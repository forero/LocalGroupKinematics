```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
filename = '../data/pairs_AbacusSummit_highbase_c000_ph100_z0.100.hdf5'
data = {}
f = h5py.File(filename, 'r')
for k in f.keys():
    data[k] = f[k][...]
f.close()
print('finished reading {}'.format(filename))
```
